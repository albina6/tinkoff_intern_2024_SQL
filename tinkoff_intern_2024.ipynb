
   
## 1 Задание
Напишите запрос, с помощью которого можно найти дубли в поле email из таблицы Sfaff.  
  
## 2 Задание  
Напишите запрос, с помощью которого можно определить возраст каждого сотрудника из таблицы Staff на момент запроса.  
  
## 3 Задание  
Напишите запрос, с помощью которого можно определить должность (Jobtitles.name) со вторым по величине уровнем зарплаты.

## 1 Задание

~~~~MySQL

SELECT email
  FROM Sfaff
 GROUP BY email
HAVING COUNT(email) > 1;
~~~~

## 2 Задание

~~~~MySQL
SELECT staff_id,
       name,
       (YEAR(CURRENT_DATE)-YEAR(birthday))-(RIGHT(CURRENT_DATE,5)<RIGHT(birthday, 5)
          ) AS age
  FROM Sfaff;
~~~~

## 3 Задание

~~~~MySQL

SET @second_salary = (
    SELECT salary
      FROM Sfaff
     GROUP BY salary
     ORDER BY salary DESC
     LIMIT 1 OFFSET 1);
     
SELECT Jobtitles.name
  FROM Sfaff
 INNER JOIN Jobtitles ON Sfaff.salary = @second_salary
                     AND Jobtitles.jobtitle_id = Sfaff.jobtitle_id;
 
~~~~
