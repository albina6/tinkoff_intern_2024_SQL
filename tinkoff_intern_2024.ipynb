## 2 

~~~~MySQL
SELECT staff_id,
       name,
       (YEAR(CURRENT_DATE)-YEAR(birthday))-(RIGHT(CURRENT_DATE,5)<RIGHT(birthday, 5)
          ) AS age
  FROM Sfaff;
~~~~

## 3 

~~~~MySQL

SET @second_salary = (
    SELECT salary
      FROM Sfaff
     GROUP BY salary
     ORDER BY salary DESC
     LIMIT 1 OFFSET 1);
     
SELECT Jobtitles.name
  FROM Sfaff
 INNER JOIN Jobtitles ON Sfaff.salary = @second_salary
                     AND Jobtitles.jobtitle_id = Sfaff.jobtitle_id;
 
~~~~
